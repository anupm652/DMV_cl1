```python
import pandas as pd
import seaborn as sns
# import numpy as np
```

```python
data = pd.read_csv('telecom_cust_chrun1.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data['Gender'] = data['Gender'].str.lower()
data['Gender']
```

```python
data['Age'] = data['Age'].astype('int')
data['Age']
```

```python
sns.pairplot(data)
```

```python
from scipy import stats
z_scores = stats.zscore(data['Monthly_charges'])
threshold = 3

# Detect outliers
outliers = data[z_scores > threshold]

# Remove outliers
df = data[z_scores <= threshold]

# Check the number of rows before and after handling outliers
print("Number of rows before handling outliers:", len(data))
print("Number of rows after handling outliers:", len(df))


```
