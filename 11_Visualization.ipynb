```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('city_day.csv')
data
```

```python
data.isnull().sum()
```

```python
# 3. Select relevant variables
date_column = "Date"
aqi_column = "AQI"
pollutant_columns = ["PM2.5", "PM10", "CO"]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data[date_column], data[aqi_column], label="AQI", color='b')
plt.xlabel("Date")
plt.ylabel("AQI")
plt.title("AQI Trend Over Time")
plt.legend()
plt.show()
```

```python
for pollutant in pollutant_columns:
    plt.figure(figsize=(12, 6))
    plt.plot(data[date_column], data[pollutant])
    plt.title(f'{pollutant} trends over time')
    plt.xlabel('Date')
    plt.ylabel(pollutant)
    plt.grid(True)
    plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(data[date_column], data[aqi_column])
plt.title("AQI Values Across Different Dates")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(data["PM2.5"], data["AQI"], alpha=0.5)
plt.title("Scatter Plot: AQI vs. PM2.5")
plt.xlabel("PM2.5")
plt.ylabel("AQI")
plt.grid(True)
plt.show()
```
